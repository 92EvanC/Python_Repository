# Merging orders_products_co

```python
#merging products with ords_prods_combined & renaming df_complete_data
df_complete_data=df_prods_perf.merge(orders_products_combined, on ='product_id', indicator=True)
```

## changing data type to conserve memory

```python
#converting categorical columns to categorical type data
df_complete_data['reordered'] = df_complete_data['reordered'].astype('category')
df_complete_data['_merge'] = df_complete_data['_merge'].astype('category')
df_complete_data['orders_day_of_week']=df_complete_data['orders_day_of_week'].astype('category')
```

```python
#changing data type to reduce memory used
df_complete_data['customer_number'] = df_complete_data['customer_number'].astype('str')
df_complete_data['order_number'] = df_complete_data['order_number'].astype('int8')
df_complete_data['orders_day_of_week'] = df_complete_data['orders_day_of_week'].astype('int8')
df_complete_data['order_hour_of_day'] = df_complete_data['order_hour_of_day'].astype('int8')
df_complete_data['product_id'] = df_complete_data['product_id'].astype('int32')
df_complete_data['aisle_id'] = df_complete_data['aisle_id'].astype('int8')
df_complete_data['department_id'] = df_complete_data['department_id'].astype('int8')
df_complete_data['days_since_prior_order'] = df_complete_data['days_since_prior_order'].astype('float16')
df_complete_data['prices'] = df_complete_data['prices'].astype('float32')
```

## Renaming 'customer_number' column to 'user_id'

```python
df_complete_data.rename(columns={'customer_number':'user_id'}, inplace=True)
```
