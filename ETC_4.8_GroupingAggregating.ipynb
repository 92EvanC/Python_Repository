## Question 6:
### Creating a new 'mean_product_price' column to use for the creation of 'spending_flag'

```python
ords_prods_merge['mean_product_price']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

#### Creating 'spending_flag'
#### average price of product purchased is < 10 = 'Low spender'
#### average price of product purchased is >= 10 = 'High spender'

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price']< 10, 'spending_flag']='Low spender'
ords_prods_merge.loc[ords_prods_merge['mean_product_price']>= 10, 'spending_flag']= 'High spender'
```

```python
#checking counts of spender type
ords_prods_merge['spending_flag'].value_counts()
```

## Question 7:
### Creating 'median_day_between_ords' column to use as customer frequency flag

```python
ords_prods_merge['median_day_between_ords']=ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

### Defining parameters for 'customer_freq_flag'
#### median of days since prior order >20 = 'Non-frequent customer'
#### median of days since prior order > 10 & <= 20 = 'Regular customer
#### median of days since prior order <= 10 = 'Frequent customer'

```python
ords_prods_merge.loc[ords_prods_merge['median_day_between_ords']>20, 'customer_freq_flag']='Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_day_between_ords']>10) & (ords_prods_merge['median_day_between_ords']<=20), 'customer_freq_flag']= 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_day_between_ords']<= 10, 'customer_freq_flag']= 'Frequent customer'
```
